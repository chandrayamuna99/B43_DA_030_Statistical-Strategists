```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("spotify_songs_dataset.csv")
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.skew(numeric_only=True)
```

```python
df["duration"] = df["duration"].fillna(df["duration"].median())
df["language"] = df["language"].fillna("Unknown")
df["collaboration"] = df["collaboration"].fillna("No collaboration")
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["release_date"]=pd.to_datetime(df["release_date"])
```

```python
df.info()
```

```python
df.head()
```

```python
df["duration_category"]=pd.cut(df["duration"],bins=[0,180,300,600],labels=["Short","Medium","Long"])
df["popularity_tier"]=pd.cut(df["popularity"],bins=[0,30,70,100],labels=["low","Medium","High"])
```

```python
df.head()
```

```python
df.to_csv("spotify_songs_cleaned_dataset.csv",index=False)
```
