```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("spotify_songs_cleaned_dataset.csv")
df.head(5)
```

```python
top_songs = df[['song_title', 'artist', 'stream']].sort_values(by='stream', ascending=False).head(5)

plt.figure(figsize=(10, 5))
plt.barh(top_songs['song_title'], top_songs['stream'], color='purple')
plt.xlabel('Streams')
plt.ylabel('Song Title')
plt.title('Top 5 Most Streamed Songs')
plt.gca().invert_yaxis()  # Invert to show highest at the top
plt.show()
```

```python
import seaborn as sns

top_genres = df.groupby('genre')['popularity'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_genres.index, y=top_genres.values, palette='coolwarm')
plt.xticks(rotation=45)
plt.xlabel('Genre')
plt.ylabel('Average Popularity')
plt.title('Top 10 Genres by Popularity')
plt.show()
```

```python
explicit_popularity = df.groupby('explicit_content')['popularity'].mean()

plt.figure(figsize=(7, 5))
sns.barplot(x=explicit_popularity.index, y=explicit_popularity.values, palette='magma')
plt.xlabel('Explicit Content')
plt.ylabel('Average Popularity')
plt.title('Explicit Content vs Popularity')
plt.show()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
df['year'] = df['release_date'].dt.year

popularity_trend = df.groupby('year')['popularity'].mean()

plt.figure(figsize=(10, 5))
plt.plot(popularity_trend.index, popularity_trend.values, marker='o', linestyle='-', color='red')
plt.xlabel('Year')
plt.ylabel('Average Popularity')
plt.title('Average Popularity Over Time')
plt.grid(True)
plt.show()
```
